Show P_Haar(F)

```python
hist = circuits.show_p_haar_f(input_qubit_number)

def on_change(value):
    #print(int(value))
    #input_qubit_number = int(value)
    f = np.linspace(0, 1, 1000)
    N = 2 ** int(4)
    hist.plot(f, (N - 1) * (1 - f) ** (N - 2))
    hist.show()
    
input_qubit_slider = widgets.IntSlider(
    value=input_qubit_number,
    min=0,
    max=25,
    step=1,
    description='input_qubit:',
)
input_qubit_slider.observe(on_change, names='value')
input_qubit_slider
```
